```python
import numpy as np
import matplotlib.pyplot as plt

iq = np.linspace(-0.5, 0.5, 100)
II, QQ = np.meshgrid(iq, iq)

r = 0.3
angles = np.array([-np.pi/4, np.pi/4, 3*np.pi/4+0.1])
peak_weight = [1.0, 0.9, 0.8]
peak_pos = r * np.array([[np.cos(ang), np.sin(ang)] for ang in angles])
sigma = 0.1

peaks = []
for peak, weight in zip(peak_pos, peak_weight):
    peaks.append(weight * np.exp(-((II - peak[0])**2 + (QQ - peak[1])**2)/(2*sigma**2)))
peaks_sum = np.sum(peaks, axis=0)
peaks_sum.shape
plt.pcolor(II, QQ, peaks_sum, shading='auto');
plt.axis('equal');
```
